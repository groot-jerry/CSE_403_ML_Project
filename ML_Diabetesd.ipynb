X Y **Split**

```python
x = df.drop(['Outcome'],axis=1)
y = df["Outcome"]
x.info()
```

Train Test **Split**

```python
x, xtest, y,   ytest = tts(x, y, test_size = 0.2, random_state=2)
```

Separate **X**

```python
x1 = x["Pregnancies"]
x2 = x["Glucose"]
x3 = x["BloodPressure"]
x4 = x["SkinThickness"]
x5 = x["Insulin"]
x6 = x["BMI"]
x7 = x["DiabetesPedigreeFunction"]
x8 = x["Age"]
```

```python
# parameter initialization
theta = [1, 1.2, 1.5, 1.3, 2, 1.7, 1.8, 2.3, 1.6]
print(theta)

J = 100
for t in range(100000):
    # hypothesis function

    h = []
    m = len(y)
    for i in range(m):
        h.append(theta[0] + theta[1]*x1.iloc[i] + theta[2]*x2.iloc[i] + theta[3]*x3.iloc[i] + theta[4]*x4.iloc[i] + theta[5]*x5.iloc[i] + theta[6]*x6.iloc[i] + theta[7]*x7.iloc[i] + theta[8]*x8.iloc[i])

    # print("Hypothesis Function Value is: ", h)

    # cost

    error = 0
    for i in range(m):
        error = error + (h[i] - y.iloc[i]) ** 2

    J = (1/ (2*m))*error

    print(f"Cost function value is {J} and iteration is {t}")


    # Gradient Descent
    sum0 = 0
    sum1 = 0
    sum2 = 0
    sum3 = 0
    sum4 = 0
    sum5 = 0
    sum6 = 0
    sum7 = 0
    sum8 = 0
    for i in range(m):
        sum0 = sum0 + (h[i] - y.iloc[i])
        sum1 = sum1 + (h[i] - y.iloc[1])*x1.iloc[i]
        sum2 = sum2 + (h[i] - y.iloc[i])*x2.iloc[i]
        sum3 = sum3 + (h[i] - y.iloc[i])*x3.iloc[i]
        sum4 = sum4 + (h[i] - y.iloc[i])*x4.iloc[i]
        sum5 = sum5 + (h[i] - y.iloc[i])*x5.iloc[i]
        sum6 = sum6 + (h[i] - y.iloc[i])*x6.iloc[i]
        sum7 = sum7 + (h[i] - y.iloc[i])*x7.iloc[i]
        sum8 = sum8 + (h[i] - y.iloc[i])*x8.iloc[i]


    alpha = 0.000001
    theta[0] = theta[0] - (alpha/m)*sum0
    theta[1] = theta[1] - (alpha/m)*sum1
    theta[2] = theta[2] - (alpha/m)*sum2
    theta[3] = theta[3] - (alpha/m)*sum3
    theta[4] = theta[4] - (alpha/m)*sum4
    theta[5] = theta[5] - (alpha/m)*sum5
    theta[6] = theta[6] - (alpha/m)*sum6
    theta[7] = theta[7] - (alpha/m)*sum7
    theta[8] = theta[8] - (alpha/m)*sum8

    # print("Updated parameters are: ", theta)

print(f"J is {J}")
```

Build the Raw **Model**

```python
def raw_linear_regression_predict(data):
  y = []
  for _,x in data.iterrows():
    y.append(theta[0] + theta[1]*x[0] + theta[2]*x[1] + theta[3]*x[2] + theta[4]*x[3] + theta[5]*x[4] + theta[6]*x[5] + theta[7]*x[6] + theta[8]*x[7])
  return pd.DataFrame(y)
```

Evaluation of Raw **model**

```python
y_pred_raw = raw_linear_regression_predict(xtest)

r2_raw = r2_score(ytest, y_pred_raw)
mae_raw = mean_absolute_error(ytest, y_pred_raw)
mse_raw = mean_squared_error(ytest, y_pred_raw)
rmse_raw = root_mean_squared_error(ytest, y_pred_raw)

print(r2_raw, mae_raw, mse_raw, rmse_raw)
```

Training the **model**

```python
lr = LinearRegression()
```

```python
lr.fit(x,y)
```

```python
ypred  = lr.predict(xtest)
```

Evaluation of SK Learn **model**

```python
r2_sk = r2_score(ytest, ypred)
mae_sk = mean_absolute_error(ytest, ypred)
mse_sk = mean_squared_error(ytest, ypred)
rmse_sk = root_mean_squared_error(ytest, ypred)

print(r2_sk, mae_sk, mse_sk, rmse_sk)
```
